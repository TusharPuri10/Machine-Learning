Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Loading the diabetes data set

```python
X, y = datasets.load_diabetes(return_X_y=True)
# The input return_X_y=True signals that X will be a data matrix, and y will be the regression target.
print(X.shape)
print(X[0])
print(X[1])
print(X[2])
```

Selecting a portion of this dataset 

```python
X = X[:, np.newaxis, 2]
print(X.shape)
print(X[0:3]) # just to show that we selected a feature column (1 independent variable)
```

Splitting the data (train/test split)(using scikit-learn)

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Loading and then Training the Model

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Creating a prediction using test data

```python
y_pred = model.predict(X_test)
```

Showing the data in a plot

```python
plt.scatter(X_test, y_test, color = 'black')
plt.plot(X_test, y_pred, color = 'blue', linewidth =3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
